```python
import omero.clients
from omero.gateway import BlitzGateway
```

```python
conn = BlitzGateway(username, password, host=host, port=port,
                    secure=True, group=group)
conn.connect()
```

```python
project = next(iter(conn.getObjects("Project", attributes={"name": project})))
```

```python
name_to_child = {c.getName():c for c in project.listChildren()}
name_to_child.keys()
```

```python
objs.keys()
```

```python
for name, obj in objs.items():
    meta = obj['metadata']
    dataset = name_to_child[name]
    image = next(iter(dataset.listChildren())) # assumes there is only 1 image per dataset
    key_value_data = [[k, v] for k, v in meta.items()]
    
    map_ann = omero.gateway.MapAnnotationWrapper(conn)
    map_ann.setNs(omero.constants.metadata.NSCLIENTMAPANNOTATION)
    map_ann.setValue(key_value_data)
    map_ann.save()
    image.linkAnnotation(map_ann)
```
